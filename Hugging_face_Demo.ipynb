NLP Processing

Sentiment Analysis

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
result = classifier("I was not happy with American Pie Movie")
print(result)
```

```python
pipeline(task = "sentiment-analysis") ("i was confused with the interstellar movie")
```

Batch Sentiment analysis

```python
classifier = pipeline(task = "sentiment-analysis", model = "s-nlp/russian_toxicity_classifier")
task_list = ["i like autoencoders.",\
             "i do not like pizza.",\
             "i love new york.",
             "i am neutral about Machine learning."]
classifier(task_list)
```

Text Generation

```python
from transformers import pipeline

text_generator = pipeline(task ="text-generation", model = "distilbert/distilgpt2")
generated_text = text_generator("today is rainy day in dallas",
                                 truncation=True,
                                 num_return_sequences = 2)
print("Generated_Text:\n", generated_text[0]['generated_text'])
```
